# Authentification

```python
# import libs needed to process data
import numpy as np
import pandas as pd
# imports libs needed to access data from google sheets
import google.colab as colab
import gspread
import google.auth as gauth

# sets up authentification
colab.auth.authenticate_user()

creds, _ = gauth.default()

gc = gspread.authorize(creds)

# setting up data table display
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
# main scouting response sheet
sheet_name = "2023 Charged Up: Capital City Classic (Responses)"
source_pg = "CCC"

sheet = gc.open(sheet_name)
source = sheet.worksheet(source_pg)
df = pd.DataFrame(source.get_all_records())

# picklist sheet
pl_sheet_name = "2023 CCC Picklist"
pl_pg = "Picklist"

picklist_fin = gc.open(pl_sheet_name)
picklist_pg = picklist_fin.worksheet(pl_pg)
```

**bold text**# Define Various Parameters for Picklisting

```python
team_list = [8,253,649,701,766,841,846,852,1072,1425,1458,1671,1678,1967,2073,2135,2288,2251,2643,3189,3257,3482,3598,3859,4135,4159,4255,4643,4698,5027,5274,5458,5817,5924,5940,6059,6918,7137,7419,7528,8016,8768,9400,9634,9916,9919,9924,9940]
```

```python
col_list = list(df.columns) # turns column indexes into lists based on gps and section
cone_auto = col_list[7:13:2]
cube_auto = col_list[8:13:2]
cone_teleop = col_list[19:25:2]
cube_teleop = col_list[20:25:2]

co_au_h = col_list[7]
co_au_m = col_list[9]
co_au_l = col_list[11]

cu_au_h = col_list[8]
cu_au_m = col_list[10]
cu_au_l = col_list[12]

co_tel_h = col_list[19]
co_tel_m = col_list[21]
co_tel_l = col_list[23]

cu_tel_h = col_list[20]
cu_tel_m = col_list[22]
cu_tel_l = col_list[24]
```

# Modifies Data

```python
# cleans up data to make all values int
for col in cone_auto+cone_teleop+cube_auto+cube_teleop:
  df[col] =  pd.to_numeric(df[col], errors='coerce', downcast="integer").fillna(0)
```

```python
# filter out team numbers that are not in final list (some scouters enter teams that dont exist)
team_num = 'Team Number (NUMBER ONLY) - ex. 2'
df=df.loc[df[team_num].isin(team_list)]

# makes list of teams
indiv_team = df.groupby(col_list[4])
team_data_list = list(filter(lambda x: isinstance(x,int), list(df[team_num].unique())))

team_data_list.sort()
```

# Processes Data

```python
stats_df = indiv_team[cone_auto + cube_auto + cone_teleop + cube_teleop].aggregate(["mean","min","max","std"]).round(2).fillna(0)
stats_df
```

```python
stats_df['Cone Auto Points Avg'] = stats_df[[(co_au_l,"mean"),(co_au_m,"mean"),(co_au_h,"mean")]].multiply([3,4,6]).sum(axis=1).round(2)
stats_df['Cone Teleop Points Avg'] = stats_df[[(co_tel_l,"mean"),(co_tel_m,"mean"),(co_tel_h,"mean")]].multiply([2,3,5]).sum(axis=1).round(2)

stats_df['Cube Auto Points Avg'] = stats_df[[(cu_au_l,"mean"),(cu_au_m,"mean"),(cu_au_h,"mean")]].multiply([3,4,6]).sum(axis=1).round(2)
stats_df['Cube Teleop Points Avg'] = stats_df[[(cu_tel_l,"mean"),(cu_tel_m,"mean"),(cu_tel_h,"mean")]].multiply([2,3,5]).sum(axis=1).round(2)

stats_df['Cone Auto Standard Deviation'] = stats_df[[(co_au_l,"std"),(co_au_m,"std"),(co_au_h,"std")]].mean(axis=1).round(2)
stats_df['Cone Teleop Standard Deviation'] = stats_df[[(co_tel_l,"std"),(co_tel_m,"std"),(co_tel_h,"std")]].mean(axis=1).round(2)

stats_df['Cube Auto Standard Deviation'] = stats_df[[(cu_au_l,"std"),(cu_au_m,"std"),(cu_au_h,"std")]].mean(axis=1).round(2)
stats_df['Cube Teleop Standard Deviation'] = stats_df[[(cu_tel_l,"std"),(cu_tel_m,"std"),(cu_tel_h,"std")]].mean(axis=1).round(2)
```

# Final Picklist

```python
picklist_df = stats_df.sort_values(by=['Cone Teleop Points Avg','Cone Teleop Standard Deviation','Cone Auto Points Avg','Cone Auto Standard Deviation'],ascending = [False,True,False,True])
final_pl_df = picklist_df[['Cone Teleop Points Avg','Cone Auto Points Avg','Cone Teleop Standard Deviation','Cone Auto Standard Deviation']]
picklist_df
```

# Prints Picklist to new sheet

```python
picklist_data = final_pl_df.reset_index().values.tolist()

picklist_pg.freeze(0,0)
picklist_pg.append_rows(picklist_data)
picklist_pg.freeze(1,1)
```
